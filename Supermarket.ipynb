```python
plt.figure()
#plt.style.use('ggplot')
group_by_city=pd.Grouper(key="city")
revenue_by_city=sm_df.groupby(group_by_city)["revenue"].sum()
colors = ['#FAC500', '#FADF6B', '#FBEBA9']
plt.bar(revenue_by_city.index, revenue_by_city, color=colors, width=0.3)
plt.xlabel("City")
plt.ylabel("Revenue")
plt.title("Revenue generated per city")
plt.show
```

```python
plt.figure(figsize=(13, 7))
group_by_product=pd.Grouper(key="product_line")
revunue_by_productline=sm_df.groupby(group_by_product)["revenue"].sum()
col=("#FCA182","#FCD982","#EBFC82","#89FC82","#82FCD9","#82DBFC","#8289FC")
plt.bar(revunue_by_productline.index, revunue_by_productline, color=col, width=0.5)
plt.xlabel("Product line")
plt.ylabel("Revenue")
plt.title("Revenue generated per product")
plt.show()
```
